1. Создайте матрицу размером 8х8 и заполните ее шахматным рисунком

````python
size = 8

# Через срезы
first_arr= np.zeros((size, size), dtype=int)
first_arr[1::2, 1::2] = 1 
first_arr[::2, ::2] = 1

# Через where
row, col = np.indices((size, size))
second_arr = np.where((row + col) % 2 == 0, 1, 0)

print('\tfirst_arr\n', first_arr, '\n\n\tsecond_arr \n', second_arr)
````

5. Как найти общие значения между двумя массивами?

````python
first_array = np.array([3, 2, 3, 4, 5])
second_array = np.array([8, 5, 3, 9, 4])

common_values = np.intersect1d(first_array, second_array)

print(common_values)

first_array = np.array([[2, 5, 6, 11], [4, 5, 6, 14]])
second_array = np.array([[1, 5, 9], [7, 8, 11]])

common_values = np.intersect1d(first_array.ravel(), second_array.ravel())

print(common_values)
````
